```python
no_pages = 1

def get_data(pageNo):  
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    r = requests.get('https://www.superseis.com.py/Ofertas.aspx')
    content = r.content
    soup = BeautifulSoup(content)
    #print(soup)

    alls = []
    for d in soup.find_all('div', attrs={'class':'item-box'}):
        #print(d)
        name = d.find('h2', attrs={'class':'product-title'})
        product_brand = d.find('div', attrs={'class':'product-brand'})
        price = d.find('div', attrs={'class':'prices'})

        all1=[]

        if name is not None:
            #print(n[0]['alt'])
            #all1.append(n[0]['alt'])
            all1.append(name.text)
        else:
            all1.append("unknown-product")

        if  product_brand is not None:
            #print(author.text)
            all1.append(product_brand.text)
        elif product_brand is None:
            product_brand = d.find('div', attrs={'class':'product-brand'})
            if product_brand is not None:
                all1.append(product_brand.text)
            else:    
                all1.append('0')    
        if price is not None:
            #print(price.text)
            all1.append(price.text)
        else:
            all1.append('0')
        alls.append(all1)    
    return alls
```

```python
results = []
for i in range(1, no_pages+1):
  results.append(get_data(i))
```

```python
results
```

```python
flatten = lambda l: [item for sublist in l for item in sublist]
df = pd.DataFrame(flatten(results),columns=['Nombre','Marca','Precio'])
df.to_csv('super_seis_ofertas.csv', index=False, encoding='utf-8')
```
